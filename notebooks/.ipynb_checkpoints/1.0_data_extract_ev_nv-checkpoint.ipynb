```python
import numpy as np
import pandas as pd
import warnings

pd.set_option('display.max_columns', 40)
warnings.filterwarnings('ignore')
```

```python
raw_538 = pd.read_csv("../data/538_presidential_national_toplines_2020.csv")

raw_economist1 = pd.read_csv("../data/economist_electoral_college_votes_over_time.csv")
raw_economist2 = pd.read_csv("../data/economist_national_predicted_popular_vote.csv")
```

```python
raw_538.shape, raw_economist1.shape, raw_economist2.shape
```

# 1. EXTRACT 538'S EV PROJECTIONS FROM JUNE 1 

```python
raw_538.head()
```

```python
raw_538['modeldate'] = pd.to_datetime(raw_538['modeldate'])
```

```python
cols1 = ["modeldate", "candidate_inc", "ev_inc", "national_voteshare_inc"]

cols2 = ["modeldate", "candidate_chal", "ev_chal", "national_voteshare_chal"]


trump_538 = raw_538[cols1].copy()
biden_538 = raw_538[cols2].copy()
```

```python
trump_538 = trump_538.rename(
    columns={
        "modeldate": "Forecast_Date",
        "candidate_inc": "Candidate",
        "ev_inc": "538's Projection of Trump's EV",
        "national_voteshare_inc": "538's Projection of Trump's National Vote Share (%)",
    }
)
```

```python
biden_538 = biden_538.rename(
    columns={
        "modeldate": "Forecast_Date",
        "candidate_chal": "Candidate",
        "ev_chal": "538's Projection of Biden's EV",
        "national_voteshare_chal": "538's Projection of Biden's National Vote Share (%)",
    }
)
```

# 2. EXTRACT ECONOMIST'S PROJECTIONS FROM JUNE 1

## 2.1 EXTRACT ECONOMIST'S EV PROJECTIONS 

```python
raw_economist1.head()
```

```python
raw_economist1['date'] = pd.to_datetime(raw_economist1['date'])
```

```python
economist_ev = raw_economist1[raw_economist1["date"] >= "2020-06-01"].sort_values(
    by="date", ascending=False
)
```

```python
economist_ev["Candidate"] = np.where(economist_ev["party"] == "democratic", "Biden", "Trump")
```

```python
cols3 = ["date", "Candidate", "median_ev"]

trump_economist_ev = economist_ev[economist_ev["Candidate"] == "Trump"][cols3].copy()
biden_economist_ev = economist_ev[economist_ev["Candidate"] == "Biden"][cols3].copy()
```

```python
trump_economist_ev = trump_economist_ev.rename(
    columns={
        "date": "Forecast_Date",
        "Candidate": "Candidate",
        "median_ev": "Economist's Projection of Trump's EV",
    }
)

biden_economist_ev = biden_economist_ev.rename(
    columns={
        "date": "Forecast_Date",
        "Candidate": "Candidate",
        "median_ev": "Economist's Projection of Biden's EV",
    }
)
```

```python
biden_economist_ev.head()
```

```python
trump_economist_ev.head()
```

```python
trump_economist_ev.shape, biden_economist_ev.shape
```

## 2.2 EXTRACT ECONOMIST'S NATIONAL VOTE SHARE PROJECTIONS 

```python
raw_economist2.head()
```

```python
raw_economist2['date'] = pd.to_datetime(raw_economist2['date'])

raw_economist2['projected_vote_mean'] = raw_economist2['projected_vote_mean'] * 100
```

```python
economist_nv = raw_economist2[raw_economist2["date"] >= "2020-06-01"].sort_values(
    by="date", ascending=False
)
```

```python
economist_nv.shape
```

```python
economist_nv.tail()
```

```python
economist_nv["Candidate"] = np.where(economist_nv["party"] == "democratic", "Biden", "Trump")
```

```python
cols4 = ["date", "Candidate", "projected_vote_mean"]

trump_economist_nv = economist_nv[economist_nv["Candidate"] == "Trump"][cols4].copy()
biden_economist_nv = economist_nv[economist_nv["Candidate"] == "Biden"][cols4].copy()
```

```python
trump_economist_nv = trump_economist_nv.rename(
    columns={
        "date": "Forecast_Date",
        "Candidate": "Candidate",
        "projected_vote_mean": "Economist's Projection of Trump's National Vote Share (%)",
    }
)

biden_economist_ev = biden_economist_ev.rename(
    columns={
        "date": "Forecast_Date",
        "Candidate": "Candidate",
        "projected_vote_mean": "Economist's Projection of Biden's National Vote Share (%)",
    }
)
```
